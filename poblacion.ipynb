```python
# Imports necesarios para la elaboraci√≥n del trabajo
import pandas as pd
import ipywidgets as widgets
from ipyleaflet import Map, MagnifyingGlass, basemaps, basemap_to_tiles, Marker
```

```python
# Datos de la poblaci√≥n desde 1960 hasta el 2020
datosPoblacion = pd.read_csv("data/urban_total.csv")
```

```python
# Datos de la Geolocalizaci√≥n de todos los paises
datosGeolocation = pd.read_csv("data/countries.csv")
```

```python
# Cambio de nombre a country code
datosPoblacion2 =  datosPoblacion.rename(columns={'Country Code':'iso3'})
```

```python
datosGeolocation.head(5)
```

```python
datosPoblacion2.head(5)
```

```python
# Uni√≥n de la data por medio de iso3
result = pd.merge(datosGeolocation, datosPoblacion2, on="iso3")
```

```python
result.keys()
```

```python
# Se filtra la data
finalData = result[["name","latitude","longitude","2020","capital","subregion","region"]]
```

```python
finalData
```

```python
#Se convierte en diccionario la dara filtrada
diccionario = finalData.to_dict('records')
diccionario
```

```python
#Convertir en tuplas
lista_tuplas = [((d['name']), (d['latitude'], d['longitude'],d['2020'],d['name'],d['capital'],d['subregion'],d['region']))
                  for d in diccionario]
lista_tuplas[0]
```

```python
# Widgets por pais
lat_long = widgets.Dropdown(
    options = lista_tuplas,
    value=(33.0, 65.0, 10131490.0, 'Afghanistan','Kabul','Southern Asia', 'Asia'),
    description= 'Pais:',
)
lat_long
```

```python
latitud = lat_long.value[0]
longitud = lat_long.value[1]
poblacion = lat_long.value[2]
nombre = lat_long.value[3]
capital = lat_long.value[4]
sub_region = lat_long.value[5]
continente = lat_long.value[6]
```

```python
# Generaci√≥n de mapa
center = (0,0)
m = Map(center=center, zoom = 2, basemap=basemaps.Stamen.Terrain)
c = (latitud,longitud)
marker = Marker(location=c, draggable = False, title=" Pa√≠s: " + nombre + "\n Poblaci√≥n: " + str(poblacion)+
                "\n Capital: " + capital +
                "\n Regi√≥n: " + continente +
                "\n Sub regi√≥n: " + sub_region, alt=nombre)
m.add_layer(marker)
                 
m
```
